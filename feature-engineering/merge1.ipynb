```python
import os
import urllib.request

# to read excel
#df = pd.read_excel(file_https://github.com/fenago/MLEssentials/blob/main/datasets/Online%20Retail.xlsx?raw=trueurl)

data_location = '../datasets/Online-Retail.csv'
data_url = 'https://s3.amazonaws.com/elephantscale-public/data/retail-data/all/online-retail-dataset.csv'
# data_url = 'https://raw.githubusercontent.com/elephantscale/datasets/master/retail-data/all/online-retail-dataset.csv'

if not os.path.exists(data_location):
    data_location = os.path.basename(data_location)
    if not os.path.exists(data_location):
        print ("Downloading  ", data_url)
        urllib.request.urlretrieve(data_url, data_location)
print ('data_location: ', data_location)
```

```python
import pandas as pd

df = pd.read_csv(data_location)
df
```

```python
uk_holidays_2010 = pd.read_csv('https://date.nager.at/PublicHoliday/Country/GB/2010/CSV')

uk_holidays_2010
```

```python
uk_holidays_2011 = pd.read_csv('https://date.nager.at/PublicHoliday/Country/GB/2011/CSV')
uk_holidays_2011
```

```python
uk_holidays = uk_holidays_2010.append(uk_holidays_2011)
uk_holidays.shape
```
